```python
import random
import matplotlib.pyplot as plt

start_node = 0
end_node = 12
num_nodes = len(adjacency_matrix)

def is_valid_path(path):
    # Verify that consecutive nodes in the path are connected in the adjacency matrix
    valid = all(adjacency_matrix[path[i]][path[i+1]] != float('inf') for i in range(len(path) - 1))
    if not valid:
        print(f"Invalid path: {path}")
    return valid

def generate_random_path():
    attempts = 0
    while attempts < 100:
        path = [start_node]
        while path[-1] != end_node:
            # Get all adjacent nodes not yet in path and not set to 'inf' (unreachable)
            adjacent = [i for i, v in enumerate(adjacency_matrix[path[-1]]) if v != float('inf') and i not in path]
            if not adjacent:
                break
            next_node = random.choice(adjacent)
            path.append(next_node)
            if next_node == end_node:
                if is_valid_path(path):
                    print(f"Path generated: {path}")
                    return path
                break
        attempts += 1
    return [start_node, end_node]

def create_initial_population(population_size):
    # Generate initial population of paths
    return [generate_random_path() for _ in range(population_size)]

def calculate_fitness(path):
    if isinstance(path, list):  # Check if path is a list
        if is_valid_path(path):
            # Fitness is the inverse of the path cost (lower cost is better)
            return sum(adjacency_matrix[path[i]][path[i+1]] for i in range(len(path) - 1))
        else:
            return float('inf')  # High penalty for invalid paths
    else:
        return path.fitness  # If path is an object with a 'fitness' attribute, return it directly

def calculate_cost(path, adjacency_matrix):
    cost = 0
    for i in range(len(path) - 1):
        current_city = path[i]
        next_city = path[i + 1]
        # Ensure cities are within adjacency matrix limits
        if current_city < len(adjacency_matrix) and next_city < len(adjacency_matrix):
            cost += adjacency_matrix[current_city][next_city]
        else:
            return float('inf')  # Infinite cost if cities are out of bounds
    return cost

def roulette_selection(population):
    # Calculate the total fitness of the population. Here, fitness is calculated inversely as the cost:
    # If the cost is not infinite (path is valid), use the inverse of the cost; otherwise, use 0 (for infinite cost).
    total_fitness = sum(1 / calculate_cost(path, adjacency_matrix) if calculate_cost(path, adjacency_matrix) != float('inf') else 0 for path in population)

    # Randomly pick a 'point' that will determine which path is selected, based on fitness proportionate selection (roulette wheel selection)
    pick = random.uniform(0, total_fitness)

    current = 0
    # Iterate over each path in the population to find which path corresponds to the random 'pick'
    for path in population:
        # Calculate the fitness of the current path as before
        fitness = 1 / calculate_cost(path, adjacency_matrix) if calculate_cost(path, adjacency_matrix) != float('inf') else 0
        current += fitness
        # If the cumulative fitness exceeds the picked value, this path is selected
        if current > pick:
            print(f"Selected by roulette: {path}")
            return path


def crossover(parent1, parent2):
# Check if the two parent paths are not the same to avoid unnecessary crossover
    if parent1 != parent2:
        c1, c2 = sorted(random.sample(range(1, min(len(parent1), len(parent2)) - 1), 2))
        child = parent1[:c1] + parent2[c1:c2] + parent1[c2:]
        print(f"Crossover between {parent1} and {parent2} at indices {c1}-{c2}: {child}")
        # After crossover, check if the child path is valid
        if not is_valid_path(child):
            # If the generated child is not valid, choose the better parent based on fitness
            child = parent1 if calculate_fitness(parent1) < calculate_fitness(parent2) else parent2
        return child
    # If parents are the same, return a copy of parent1
    return parent1.copy()

def mutate(path, mutation_rate=0.1):
    for _ in range(5):  # Try up to 5 mutations to find a valid path
        if random.random() < mutation_rate:
            i, j = random.randint(1, len(path)-2), random.randint(1, len(path)-2)
            # Swap nodes if it does not create an invalid path
            if all(adjacency_matrix[path[k-1]][path[l]] != float('inf') for k, l in [(i-1, j), (j, (i % (len(path)-1)) + 1), (j-1, i), (i, (j % (len(path)-1)) + 1)]):
                path[i], path[j] = path[j], path[i]
                print(f"Successful mutation: {path}")
                break
    if not is_valid_path(path):
        print("Mutation failed, invalid path generated.")
    return path

def new_generation(population, mutation_rate):
    # Use roulette selection to choose parents for the next generation
    intermediate_population = [roulette_selection(population) for _ in range(len(population))]

    # Perform crossover on selected parents to create new paths (children)
    children = [crossover(intermediate_population[i], intermediate_population[(i+1) % len(population)]) for i in range(len(population))]

    # Apply mutations to the children to introduce variation
    new_pop = [mutate(child, mutation_rate) for child in children]

    # Output the fitness of the new generation to monitor progress
    print(f"New generation created: {[calculate_fitness(ind) for ind in new_pop]}")
    return new_pop


def genetic_algorithm(population_size, generations, mutation_rate):
    # Initialize the population by creating an initial set of paths
    population = create_initial_population(population_size)
    cost_history = []  # To track the minimum cost in each generation

    # Iterate over each generation
    for i in range(generations):
        print(f"--- Generation {i+1} ---")
        # Create a new generation by selecting, crossing over, and mutating the current population
        population = new_generation(population, mutation_rate)

        # Calculate the fitness for each path in the population. Lower fitness (cost) is better.
        costs = [calculate_fitness(path) for path in population]
        # Determine the path with the best (lowest) cost in the current generation
        best_cost = min(costs)
        cost_history.append(best_cost)  # Append the best cost of this generation to the history
        print(f"Generation {i+1}: Best cost = {best_cost}")

    # Plot the evolution of the best cost over all generations
    plt.plot(cost_history)
    plt.ylabel('Cost of Best Path')
    plt.xlabel('Generation')
    plt.title('Progress of Best Path Cost per Generation')
    plt.show()

    # Find the overall best path from the final generation based on fitness
    best_path = min(population, key=lambda x: calculate_fitness(x))
    return best_path, calculate_fitness(best_path)  # Return the best path and its cost


# Genetic Algorithm parameters
population_size = 5
generations = 5
mutation_rate = 0.5  # Increase mutation rate to improve diversity

best_path, cost = genetic_algorithm(population_size, generations, mutation_rate)
print("Best path found:", best_path)
print("Cost of best path:", cost)
```
